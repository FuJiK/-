## 素数を見つけるアルゴリズム

- 一般化
    - 素数を格納するnumbersという空のリストで初期化
    - iを1～10000まで繰り返す
        - 素数判定フラグをTrueに初期化
        - 2,...,i-1の数値までで割り切れたら、FlagをFalse
        - iより小さな全ての値（1は除く）で計算して、フラグがTrueのままであれば素数として判定
        - 素数なら、配列（numbers）に追加
    - この中ではまだ２つ粗がある


```python
# 空のリストの初期化
numbers = []

# 繰り返し
for i in range(1,11):
    # 素数判定フラグをTrueに初期化
    flag = True
    # 割って、あまりの確認
    for j in range(2,i): # i の一つ手前まで
        if i % j == 0:
            flag = False
    # 素数の場合
    if flag == True:
        numbers.append(i)
```

```python
# 関数化
def get_prime_numbers(max_val):
    # 空のリストの初期化
    numbers = []

    # 繰り返し
    for i in range(1,max_val+1):
        # 素数判定フラグをTrueに初期化
        flag = True
        # 割って、あまりの確認
        for j in range(2,i): # i の一つ手前まで
            if i % j == 0:
                flag = False
        # 素数の場合
        if flag == True:
            numbers.append(i)
    return numbers
```

```python
def get_prime_numbers_2(max_val):
    # 空のリストの初期化
    numbers = []

    # 繰り返し
    for i in range(1,max_val+1):
        # 素数判定フラグをTrueに初期化
        flag = True
        # 偶数なら、False
        if i % 2 == 0:
            flag = False
        # 割って、あまりの確認
        for j in range(2,i): # i の一つ手前まで
            if i % j == 0:
                flag = False
        # 素数の場合
        if flag == True:
            numbers.append(i)
    return numbers
```

```python
%%timeit
import math

# 素数判定関数
def isPrime(num):
    # 2未満の数字は素数ではない
    if num < 2: return False
    # 2は素数
    elif num == 2: return True
    # 偶数は素数ではない
    elif num % 2 == 0: return False

    # 3 ~ numまでループし、途中で割り切れる数があるか検索
    # 途中で割り切れる場合は素数ではない
    for i in range(3, math.floor(math.sqrt(num))+1, 2):
        if num % i == 0:
            return False

    # 素数
    return True

# 素数判定
def callIsPrime(input_num=10000):
    numbers = []
    # ループしながら素数を検索する
    for i in range(1, input_num):
        if isPrime(i):
            numbers.append(i)

    # 素数配列を返す
    return numbers
```

```python
# 関数の呼び出し
print(callIsPrime(10000))
```

```python
# 修正前
def get_prime_numbers(max_val):
    numbers = []
    for i in range(1,max_val+1):
        flag = True
        for j in range(2,i): 
            if i % j == 0:
                flag = False
                
                
        if flag == True:
            numbers.append(i)
    return numbers
```

```python
# 修正後　高速化1
# break を追加した処理
def get_prime_numbers3(max_val):
    numbers = []
    for i in range(1,max_val+1):
        flag = True
        for j in range(2,i): 
            if i % j == 0:
                flag = False
                break
                
        if flag == True:
            numbers.append(i)
    return numbers
```

```python
def encode(memo):
    tmp = memo[0]
    count = 1
    ans = ""
    # 処理
    for i in range(1,len(memo)):
        if tmp == memo[i]:
            count += 1
        # elifの前の結果を出力し、countが1のときは添字をつけずに出力
        elif count == 1:
            ans += tmp
            tmp = memo[i]
        elif count >= 10:
            ans += tmp + str(count) + str(",")
            tmp = memo[i]
            count = 1
        # elseの前の結果を出力,処理が終わるとcountを1に戻す
        else:
            ans += tmp + str(count)
            tmp = memo[i]
            count = 1
    ans += tmp +str(count)
    return ans
```
